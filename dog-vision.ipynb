# Dog Vision

#### This uses the models built in `End-to-end Multil-class Dog Breed Classification` to predict the breeds of the dogs of any given picture.

* Takes the folder path in which the image/images are present and model path for the model to be used for prediction as input and predicts the breed of the dog with image and labels

### Details about the model used:
1. The model is build using transfer learning.
2. Hence the first layer choosed is - https://tfhub.dev/google/imagenet/inception_resnet_v2/classification/4, parameters for the layers in this model is frozen
3. The next non frozen layer consists if 520 neurons/kernels units and has:
    * ReLU activation
    * L1 regularization with constant 0.01 for kernal outputs
    * L2 regularization with constant 0.01 for activation outputs
    * These are followed by a softmax layer of 120 kernels wrt 120 different breeds model was trained with
4. Model is compiled using categorical cross-entrop as the loss function and Adam-optimizer having accuracy as the only metrics
